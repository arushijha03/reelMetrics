```python
import pandas as pd

merged_df = pd.read_csv("/content/Cleaned_Movies_Data_1995_2024.csv")
merged_df.head()
```

```python
import pandas as pd

def classify_movie(row):
    # Calculate ROI
    if row['BUDGET'] > 0:
        roi = (row['GROSSWORLDWIDE'] - row['BUDGET']) / row['BUDGET']
    else:
        roi = -1

    if roi >= 1.5 or row['RATING'] >= 7.5 or row['OSCARS'] >= 1:
        return "Successful"
    elif 0 <= roi < 1.5 and 5.5 <= row['RATING'] < 7.5:
        return "Mediocre"
    else:
        return "Failure"


merged_df['MOVIE_CATEGORY'] = merged_df.apply(classify_movie, axis=1)
merged_df[['TITLE', 'BUDGET', 'GROSSWORLDWIDE', 'RATING', 'OSCARS', 'MOVIE_CATEGORY']].head()
```

## PCA

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt

numerical_columns = ['DURATION', 'RATING', 'VOTES', 'BUDGET', 'GROSSWORLDWIDE', 'NOMINATIONS', 'OSCARS']
pca_ready_df = merged_df[numerical_columns]

# Standardizing the numerical data using Sklearn StandardScaler
scaler = StandardScaler()
pca_ready_scaled = scaler.fit_transform(pca_ready_df)

# Perform PCA with n_components=2 and n_components=3
pca_2d = PCA(n_components=2)
pca_2d_result = pca_2d.fit_transform(pca_ready_scaled)

pca_3d = PCA(n_components=3)
pca_3d_result = pca_3d.fit_transform(pca_ready_scaled)

# Compute explained variance ratio for 2D and 3D datasets
explained_variance_2d = np.sum(pca_2d.explained_variance_ratio_) * 100
explained_variance_3d = np.sum(pca_3d.explained_variance_ratio_) * 100

# Visualizing the 2D PCA results
plt.figure(figsize=(9, 7))
plt.scatter(pca_2d_result[:, 0], pca_2d_result[:, 1], alpha=0.5, c='blue', edgecolors='k')
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("2D Visualization of PCA (n_components=2)")
plt.grid(True)
plt.show()

# Visualizing the 3D PCA results
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_3d_result[:, 0], pca_3d_result[:, 1], pca_3d_result[:, 2], alpha=0.5, c='red', edgecolors='k')
ax.set_xlabel("Principal Component 1")
ax.set_ylabel("Principal Component 2")
ax.set_zlabel("Principal Component 3")
ax.set_title("3D Visualization of PCA (n_components=3)")
plt.show()

print(f"The variance that is explained by PCA with 2 components is: {explained_variance_2d}")
print(f"The variance that is explained by PCA with 3 components is: {explained_variance_3d}")
```

```python
#Perform full PCA for 95% variance
pca_full = PCA()
pca_transformed = pca_full.fit_transform(pca_ready_scaled)
explained_variance_ratio = np.cumsum(pca_full.explained_variance_ratio_)
num_components_95 = np.argmax(explained_variance_ratio >= 0.95) + 1

pca_df = pd.DataFrame(pca_transformed[:, :6], columns=[f"PC{i+1}" for i in range(6)])

top_3_eigenvalues = pca_full.explained_variance_[:3]

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio * 100, marker='o', linestyle='--', color='b')
plt.axhline(y=95, color='r', linestyle='--', label="95% Variance Threshold")
plt.axvline(x=num_components_95, color='g', linestyle='--', label=f"Components Needed: {num_components_95}")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance (%)")
plt.title("PCA Variance Retention and 95% Threshold")
plt.legend()
plt.grid(True)
plt.show()

print(f"The number of components needed for 95% variance: {num_components_95}")
print(f"The Top 3 eigenvalues are: {top_3_eigenvalues}")
```

## CLUSTERING

```python
#Separate label Movie Category
labels = merged_df['MOVIE_CATEGORY']
data_without_labels = merged_df.drop(columns=['MOVIE_CATEGORY'])

#Prepping data for clustering
numerical_columns = ['DURATION', 'RATING', 'VOTES', 'BUDGET', 'GROSSWORLDWIDE', 'NOMINATIONS', 'OSCARS']
quantitative_data = data_without_labels[numerical_columns]

#Normalizing the data using StandardScaler (mean=0, stddev=1)
scaler = StandardScaler()
normalized_data = scaler.fit_transform(quantitative_data)

#PCA to reduce to 3 dimensions
pca_3d = PCA(n_components=3)
pca_3d_result = pca_3d.fit_transform(normalized_data)

explained_variance_3d = np.sum(pca_3d.explained_variance_ratio_) * 100

pca_3d_df = pd.DataFrame(pca_3d_result, columns=['PC1', 'PC2', 'PC3'])

df_before_clustering = pd.DataFrame(pca_3d_df, columns=["PC1", "PC2", "PC3"])
display(df_before_clustering.head())

# Show percentage of variance retained in 3D
print(f"Variance explained by PCA with 3 components: {explained_variance_3d}")
```

```python
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch


#Finding optimal K value using the Silhouette Method
silhouette_scores = []
k_values = range(2, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(pca_3d_result)
    silhouette_avg = silhouette_score(pca_3d_result, cluster_labels)
    silhouette_scores.append(silhouette_avg)

#Selecting top three optimal K values
optimal_k_values = sorted(zip(k_values, silhouette_scores), key=lambda x: x[1], reverse=True)[:3]
optimal_k_values = [k[0] for k in optimal_k_values]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

for idx, k in enumerate(optimal_k_values):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    clusters = kmeans.fit_predict(pca_3d_result)

    ax = axes[idx]
    scatter = ax.scatter(
        pca_3d_result[:, 0], pca_3d_result[:, 1], c=clusters, cmap="viridis", alpha=0.6, edgecolors="k"
    )

    #Plotting centroids
    centroids = kmeans.cluster_centers_
    ax.scatter(centroids[:, 0], centroids[:, 1], marker="X", s=200, c="red", label="Centroids")

    ax.set_xlabel("PC1")
    ax.set_ylabel("PC2")
    ax.set_title(f"KMeans Clustering (k={k})")
    ax.legend()

plt.show()

df_after_clustering = pd.DataFrame(pca_3d_result, columns=["PC1", "PC2", "PC3"])
df_after_clustering["Cluster"] = clusters
display(df_after_clustering.head())

print(f"Optimal k values are: {optimal_k_values}")
print(f"Silhouette scores are: {silhouette_scores}")
```

```python
from sklearn.cluster import AgglomerativeClustering
plt.figure(figsize=(16, 8))

#Hierarchical clustering using Ward's method
linked = sch.linkage(pca_3d_result, method='ward')

# Generating the dendrogram
plt.figure(figsize=(16, 8))
dendrogram = sch.dendrogram(
    linked,
    truncate_mode="lastp",
    p=30,
    leaf_rotation=90,
    leaf_font_size=10,
    above_threshold_color='blue',
    color_threshold=linked[-10, 2],
    show_leaf_counts=True,
)

plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Sample Index or Clustered Points")
plt.ylabel("Distance")
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.6)

num_clusters = 5
hierarchical_clustering = AgglomerativeClustering(n_clusters=num_clusters, linkage="ward")
clusters_hierarchical = hierarchical_clustering.fit_predict(pca_3d_result)

df_after_hierarchical = pd.DataFrame(pca_3d_result, columns=["PC1", "PC2", "PC3"])
df_after_hierarchical["Cluster"] = clusters_hierarchical

print(df_after_hierarchical.head())
```

```python
#Performing DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(pca_3d_result)

df_after_dbscan = pd.DataFrame(pca_3d_result, columns=["PC1", "PC2", "PC3"])
df_after_dbscan["Cluster"] = dbscan_labels

display(df_after_dbscan.head())

plt.figure(figsize=(8, 6))
plt.scatter(pca_3d_result[:, 0], pca_3d_result[:, 1], c=dbscan_labels, cmap='plasma', alpha=0.6, edgecolors='k')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("DBSCAN Clustering Results")
plt.show()
```

```python
import ast
#Select categorical columns for transaction-based format
transactional_columns = ["GENRES", "MPA", "DIRECTORS"]

print("Columns of dataframe before being converted to transactional data:")
display(merged_df[transactional_columns].head())
```

```python
# Define the general list of movie genres
movie_genres = [
    "Action", "Adventure", "Animation", "Biography", "Comedy", "Crime", "Documentary",
    "Drama", "Family", "Fantasy", "History", "Horror", "Music", "Musical", "Mystery",
    "Romance", "Sci-Fi", "Sport", "Thriller", "War", "Western", "Superhero", "Noir",
    "Psychological", "Suspense", "Dark Comedy", "Heist", "Survival", "Martial Arts"
]

# Function to clean and convert record data to transaction format with limited, ordered genres and directors
def clean_and_convert_to_transaction_ordered(row):
    transaction = []

    # Add only up to 2 genres that exist in movie_genres list, sorted alphabetically
    try:
        genres = ast.literal_eval(row["GENRES"]) if isinstance(row["GENRES"], str) else row["GENRES"]
        filtered_genres = sorted([genre.strip().replace("'", "").replace('"', '') for genre in genres if genre in movie_genres])[:2]
        transaction.extend(filtered_genres)  # Add genres first
    except:
        pass

    # Add only up to 2 directors, sorted alphabetically
    try:
        directors = ast.literal_eval(row["DIRECTORS"]) if isinstance(row["DIRECTORS"], str) else row["DIRECTORS"]
        filtered_directors = sorted([director.strip().replace("'", "").replace('"', '') for director in directors])#[:2]
        transaction.extend(filtered_directors)  # Add directors after genres
    except:
        pass

    if pd.notna(row["MPA"]):
        transaction.append(row["MPA"].strip())

    return transaction


transactional_data_ordered = merged_df.apply(clean_and_convert_to_transaction_ordered, axis=1)

transaction_list = transactional_data_ordered.tolist()

for i in range(5):
    print(f"Transaction {i+1}: {transaction_list[i]}")
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

#Converting transactional data into the required format
te = TransactionEncoder()
te_ary = te.fit(transaction_list).transform(transaction_list)
df_transactions = pd.DataFrame(te_ary, columns=te.columns_)

#Applying the Apriori algorithm
frequent_itemsets = apriori(df_transactions, min_support=0.01, use_colnames=True)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.6)

rules['antecedents'] = rules['antecedents'].apply(lambda x: ', '.join(sorted(x)))
rules['consequents'] = rules['consequents'].apply(lambda x: ', '.join(sorted(x)))

rules['rule_pair'] = rules.apply(lambda row: tuple(sorted([row['antecedents'], row['consequents']])), axis=1)

#Removing duplicate rules where antecedents and consequents are reversed
rules = rules.drop_duplicates(subset=['rule_pair'])

rules = rules.drop(columns=['rule_pair'])

print(rules.head(5))
print(len(rules))
```

```python
# Sort rules by Support, Confidence, and Lift
top_support_rules = rules.nlargest(15, 'support')
top_confidence_rules = rules.nlargest(15, 'confidence')
top_lift_rules = rules.nlargest(15, 'lift')

print("Top 15 Rules by Support:")
display(top_support_rules)

print("\nTop 15 Rules by Confidence:")
display(top_confidence_rules)

print("\nTop 15 Rules by Lift:")
display(top_lift_rules)
```

```python
import networkx as nx
import plotly.graph_objects as go

def plot_association_network_plotly(rules_subset, title):
    G = nx.DiGraph()

    for _, row in rules_subset.iterrows():
        antecedents = row['antecedents']
        consequents = row['consequents']
        confidence = row['confidence']

        G.add_edge(antecedents, consequents, weight=confidence)

    pos = nx.spring_layout(G, seed=42)

    edge_x = []
    edge_y = []
    for edge in G.edges():
        x0, y0 = pos[edge[0]]
        x1, y1 = pos[edge[1]]
        edge_x.extend([x0, x1, None])
        edge_y.extend([y0, y1, None])

    edge_trace = go.Scatter(
        x=edge_x, y=edge_y,
        line=dict(width=1, color='royalblue'),
        hoverinfo='none',
        mode='lines'
    )

    node_x = []
    node_y = []
    node_text = []

    for node in G.nodes():
        x, y = pos[node]
        node_x.append(x)
        node_y.append(y)
        node_text.append(f"{node}")

    node_trace = go.Scatter(
        x=node_x, y=node_y,
        mode='markers+text',
        marker=dict(size=10, color='lightblue', line=dict(width=2, color='black')),
        text=node_text,
        textposition="top center",
        hoverinfo='text'
    )

    fig = go.Figure(data=[edge_trace, node_trace],
                    layout=go.Layout(
                        title=title,
                        showlegend=False,
                        hovermode='closest',
                        margin=dict(b=20, l=5, r=5, t=40),
                        xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
                        yaxis=dict(showgrid=False, zeroline=False, showticklabels=False)
                    ))

    fig.show()

plot_association_network_plotly(top_lift_rules, "Association Network for Top 15 Lift Rules (Plotly)")
```
